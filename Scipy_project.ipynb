```python
import scipy
from scipy import special

fact = special.factorial(5)
print(f"Factorial of 5: {fact}")

fact = special.factorial(6)
print(f"Factorial of 5: {fact}")
```

```python
import numpy as np
from scipy import linalg

A = np.array([[3, 2], [1, 4]])

det = linalg.det(A)
print(f"Determinant of matrix A: {det}")

b = np.array([5, 6])
x = linalg.solve(A, b)
print(f"Solution of linear system: {x}")
```

```python
from scipy import optimize

def f(x):
    return x**2 + 5*np.sin(x)

result = optimize.minimize(f, x0=0) 
print(f"Minimum value: {result.x}")
```

```python
from scipy import integrate

def f(x):
    return x**2

result, error = integrate.quad(f, 0, 3)
print("Integral of x^2 from 0 to 3:", result)
```

```python
import numpy as np
from scipy import stats

data = np.random.normal(0, 1, 1000)

mean = np.mean(data)
variance = np.var(data)
skewness = stats.skew(data)
kurtosis = stats.kurtosis(data)

print("Mean:", mean)
print("Variance:", variance)
print("Skewness:", skewness)
print("Kurtosis:", kurtosis)
```

```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 500, endpoint=False)
signal_wave = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(t.size)

b, a = signal.butter(3, 0.05)
filtered_signal = signal.filtfilt(b, a, signal_wave)

plt.plot(t, signal_wave, label='Noisy Signal')
plt.plot(t, filtered_signal, label='Filtered Signal', linestyle='--')
plt.legend()
plt.show()
```

```python
from scipy import fft
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 400)
signal_wave = np.sin(50 * 2 * np.pi * t)

signal_fft = fft.fft(signal_wave)

plt.plot(np.abs(signal_fft))
plt.title("Fourier Transform of the Signal")
plt.show()
```
